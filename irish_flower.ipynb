```python
from sklearn.datasets import load_iris  
from sklearn.model_selection import train_test_split  
from sklearn.neighbors import KNeighborsClassifier  
import numpy as np  
```

```python
knn = KNeighborsClassifier(n_neighbors=10,metric='minkowski')  # komsu sayisini artirdim minkowski formulunu kullandÄ±m
iris_veri_seti= load_iris()  
```

```python
ozellikler_egitim, ozellikler_test, hedef_egitim, hedef_test = train_test_split(iris_veri_seti['data'], iris_veri_seti['target'],random_state=0)
```

```python
# Normal veri 
ozellikler_egitim
```

```python
# Veriyi olceklendir
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
ozellikler_egitim = scaler.fit_transform(ozellikler_egitim)
ozellikler_test = scaler.transform(ozellikler_test)
ozellikler_egitim
```

```python
# Modeli eÄitiyoruz
knn.fit(ozellikler_egitim, hedef_egitim)

yeni_veri_noktasi = [[3.5, 2.1, 3.4, 1.2]]
tahmin = knn.predict(yeni_veri_noktasi)

print(tahmin)
```

```python
# Error
dogruluk = np.mean(knn.predict(ozellikler_test) == hedef_test) * 100

print("DoÄruluk:", dogruluk)
```
